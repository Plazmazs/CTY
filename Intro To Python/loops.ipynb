```python
w= int(input('width'))
h= int(input('height'))
m= h/2 -1
x= int(0)
y= 0
while x< m:
    print(" " * (w-x), "*" *(1+2*x)," "*(w-x))
    x += 1
    y += 1
while y< h:
    print(" " * (w-x), "*" *(1+2*x)," "*(w-x))
    x -= 1
    y+=1
```
